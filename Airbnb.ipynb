```python
import pandas as pd
import tkinter
import matplotlib
import matplotlib.pyplot as plt
matplotlib.use('TkAgg')
import sklearn
import csv
import math
import numpy as np
import dtale
import requests
from matplotlib import style
```

```python
df = pd.read_csv(r'Air_bnb.csv')
df[0:]
```

```python
print(df.columns)
```

```python
df.isnull().sum()
```

```python
df2 = df.fillna(0) #Replace null value to zero
df2
```

```python
df2.isnull().sum()
```

```python
'price' in df2.columns  
df2 = df[df['price']>=1000]
df_high = df2[['name','price',]]
high = pd.DataFrame.rename(df_high, columns={"name" : "House Name", "price":"Price $"})
high
```

```python

high.describe()
```

```python
'price' in df2.columns  #Filtering price below $1000
df2 = df[df['price']<=1000]
df_low = df2[['name','price']]
low = pd.DataFrame.rename(df_low, columns={"name" : "House Name", "price":"Price $"})
low
```

```python
low.describe()
```

```python
high.to_csv(r'Air_bnb_high.csv')
low.to_csv(r'Air_bnb_low.csv')
```

```python
df1=pd.read_csv(r'Air_bnb_high.csv')
graph1 = df1[['House Name','Price $']]
graph1
```

```python
ax = graph1.plot(kind='bar', title ="Airbnb",figsize=(15,10),legend=True, fontsize=12)
ax.set_xlabel("House Name",fontsize=12)
ax.set_ylabel("Price $",fontsize=12)
plt.show()
```
